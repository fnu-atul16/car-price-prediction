```python
# importing libraries
import pandas as pd
import numpy as np
```

```python
#load file
cars_df = pd.read_csv("./data/raw/cars.csv")
df = cars_df.copy( )


```

```python
missing_values_sum = cars_df.isna().sum(axis=1)
df_nan = cars_df.loc[missing_values_sum == cars_df.shape[1]-1]
index_list = df_nan.index.to_list()

cars_df.drop(index=index_list,inplace=True)

```

```python
df["car_year"] = df["name"].str.extract(r"\b(\d{4})\b")    
```

```python
df["name"] = df["name"].str.replace(r"\b(\d{4})\b","",regex=True)
```

```python
df["car_brand"] = df["name"].str.split().str[0]
```

```python
df["name"] = df["name"].str.split().str[1:].str.join(" ")

```

```python
df["name"] = df["name"].str.strip().str.upper()
```
