```python
import re
def extract_model_and_engine(df,word_dict):
    model_exact = set(word_dict.get("model",{}).get("exact",[]))
   

    model_list = []
    
    
    for val in df["name"].str.split():
        model_row = "UNKNOWN"

        for word in val:
            if word in model_exact:
                model_row = word
                break

        model_list.append(model_row)
        


    df["model"] = model_list
   


    return df
```

```python
df = extract_model_and_engine(df,pattern_dict)
```

```python

def extract_model_only(df):
    new_model = []
    new_name = []

    for name,model in zip(df["name"],df["model"]):
        tokens = str(name).split()

        if not tokens:
            new_model.append(model)
            new_name.append(name)
            continue

        first_word = tokens[0]

        if model!="UNKNOWN":
            new_model.append(model)
            remaining = tokens[1:]
        else:
            new_model.append(first_word)
            remaining = tokens[1:]


        new_name.append(" ".join(remaining))
    df["name"] = new_name
    df["model"] = new_model

    return df

```

```python
dict = {
    "Trim" : ["BASE","PREMIUM","LIMITED","LUXURY","SPORT","PLUS","S","SE","XLT","PACKAGE","LE","PLATINUM","OVERLAND","LUXE","TOURING","PREMIERE","TECHNOLOGY	"],
    "BADGE": ["AWD","LT","4MATIC","AMG","XDRIVE"]
}
```

```python
trim_list = dict["Trim"]
badge_list = dict["BADGE"]
pattern_trim = "|".join(trim_list)
pattern_badge = "|".join(badge_list)

df["Trim"] = df["name"].str.extract(f"({pattern_trim})",expand=False).fillna("UNKNOWN")
df["Badge"] = df["name"].str.extract(f"({pattern_badge})",expand =False).fillna("UNKNOWN")
```

```python
import re
temp_set =set(df["Trim"]).union(set(df["Badge"]))
temp_set.discard("UNKNOWN")
pattern = r"\b(?:{})\b".format("|".join(re.escape(x) for x in temp_set))
df["name"] = df["name"].str.replace(pattern," ",regex=True)
df["name"] = (
    df["name"]
    .astype(str)
    .str.replace(r"\s+", " ", regex=True)).str.strip()





```

```python
#manually cleaning by observation
df["model"].iloc[16] = "COOPER"
df["model"].iloc[886] = "135I"
df["model"].iloc[126] = "XDRIVE60"
```

```python
df["MPG"] = df["MPG"].astype(str).str.replace("–", "-", regex=False)
```

```python
mask = df["MPG"].str.contains(r"^\s*\d+(?:\.\d+)?\s*-\s*\d+(?:\.\d+)?\s*$"
, na=False)
df.loc[mask,"MPG"] = df.loc[mask,"MPG"].apply(lambda x: float(x.split("-")[0]) + float(x.split("-")[1])/2)
```

```python
mask2 = df["MPG"].str.contains("AUTOMATIC",na=False,case=False)
df.loc[mask2,"Transmission"] = df.loc[mask2,"MPG"]
```

```python

x = df["price"].str.strip().str.split().str[0].str.contains(r"^\$\d+,\d+$",regex = True,na=False)
df["Actual_price"] = pd.NA
df.loc[x,"Actual_price"] = df.loc[x,"price"].str.split().str[0].str.split("$").str[1].str.replace(",","").astype("Int64")

```

```python
df["Reduced_price"] = df["price"].str.extract(r"(\$[\d,]+)(?=\s*price)",expand=False)
```

```python
x = df["Reduced_price"].notna()
df.loc[x,"Reduced_price"] = df.loc[x,"Reduced_price"].str.split("$").str[1].str.replace(",","").astype(int)
```
